```python
# ============================================================
# Gold: dim_customers
# ============================================================
import polars as pl
import os

customers_silver_path = f"{SILVER_ROOT}/customers/olist_customers_silver.parquet"
df_customers_silver = pl.read_parquet(customers_silver_path)

dim_customers = df_customers_silver.select([
    "customer_id",
    "customer_unique_id",
    "customer_city",
    "customer_state",
])

# Guardar en Gold
gold_dim_customers_path = f"{GOLD_ROOT}/dim_customers/dim_customers.parquet"
os.makedirs(os.path.dirname(gold_dim_customers_path), exist_ok=True)
dim_customers.write_parquet(gold_dim_customers_path)

print(f" dim_customers guardado.")
```

```python
# ============================================================
# Gold: dim_sellers
# ============================================================
sellers_silver_path = f"{SILVER_ROOT}/sellers/olist_sellers_silver.parquet"
df_sellers_silver = pl.read_parquet(sellers_silver_path)

dim_sellers = df_sellers_silver.select([
    "seller_id",
    "seller_city",
    "seller_state",
])

gold_dim_sellers_path = f"{GOLD_ROOT}/dim_sellers/dim_sellers.parquet"
os.makedirs(os.path.dirname(gold_dim_sellers_path), exist_ok=True)
dim_sellers.write_parquet(gold_dim_sellers_path)

print(f"✅ dim_sellers guardado.")
print(f"   Filas:    {len(dim_sellers):,}")
print(f"   Columnas: {len(dim_sellers.columns)}")
```

```python
# ============================================================
# Gold: dim_products
# ============================================================
products_silver_path = f"{SILVER_ROOT}/products/olist_products_silver.parquet"
df_products_silver = pl.read_parquet(products_silver_path)

dim_products = df_products_silver.select([
    "product_id",
    "product_category_name",
])

gold_dim_products_path = f"{GOLD_ROOT}/dim_products/dim_products.parquet"
os.makedirs(os.path.dirname(gold_dim_products_path), exist_ok=True)
dim_products.write_parquet(gold_dim_products_path)

print(f"   dim_products guardado.")
print(f"   Filas:    {len(dim_products):,}")
print(f"   Columnas: {len(dim_products.columns)}")
```

```python
# ============================================================
# Gold: dim_payment
# ============================================================
payments_silver_path = f"{SILVER_ROOT}/order_payments/olist_order_payments_silver.parquet"
df_payments_silver = pl.read_parquet(payments_silver_path)

dim_payment = df_payments_silver.select([
    "order_id",
    "payment_type",
]).unique(subset=["order_id"], keep="first")  # un registro por orden

gold_dim_payment_path = f"{GOLD_ROOT}/dim_payment/dim_payment.parquet"
os.makedirs(os.path.dirname(gold_dim_payment_path), exist_ok=True)
dim_payment.write_parquet(gold_dim_payment_path)

print(f"   dim_payment guardado.")
print(f"   Filas:    {len(dim_payment):,}")
print(f"   Columnas: {len(dim_payment.columns)}")
```

```python
# ============================================================
# Gold: dim_date
# ============================================================
orders_silver_path = f"{SILVER_ROOT}/orders/olist_orders_silver.parquet"
df_orders_silver = pl.read_parquet(orders_silver_path)

dim_date = df_orders_silver.select([
    pl.col("order_purchase_timestamp").cast(pl.Date).alias("date")
]).unique().sort("date").with_columns([
    pl.col("date").dt.year().alias("year"),
    pl.col("date").dt.month().alias("month"),
    pl.col("date").dt.day().alias("day"),
    pl.col("date").dt.weekday().alias("weekday"),
])

gold_dim_date_path = f"{GOLD_ROOT}/dim_date/dim_date.parquet"
os.makedirs(os.path.dirname(gold_dim_date_path), exist_ok=True)
dim_date.write_parquet(gold_dim_date_path)

print(f"   dim_date guardado.")
print(f"   Filas:    {len(dim_date):,}")
print(f"   Columnas: {len(dim_date.columns)}")
```
