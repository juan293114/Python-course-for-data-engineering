```python
import kagglehub

# Descarga a la caché temporal
kaggle_path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")
print("Descargado en:", kaggle_path)

# Mapeo: archivo Kaggle → carpeta RAW destino
file_map = {
    "olist_orders_dataset.csv":                  RAW_ORDERS_ROOT,
    "olist_order_items_dataset.csv":             RAW_ORDER_ITEMS_ROOT,
    "olist_order_payments_dataset.csv":          RAW_ORDER_PAYMENTS_ROOT,
    "olist_order_reviews_dataset.csv":           RAW_ORDER_REVIEWS_ROOT,
    "olist_customers_dataset.csv":               RAW_CUSTOMERS_ROOT,
    "olist_products_dataset.csv":                RAW_PRODUCTS_ROOT,
    "olist_sellers_dataset.csv":                 RAW_SELLERS_ROOT,
    "olist_geolocation_dataset.csv":             RAW_GEOLOCATION_ROOT,
    "product_category_name_translation.csv":     RAW_PRODUCT_CATEGORY_ROOT,
}

for filename, dest_dir in file_map.items():
    src = os.path.join(kaggle_path, filename)
    dst = os.path.join(dest_dir, filename)
    shutil.copy2(src, dst)
    print(f"✓ {filename} → {dest_dir}")

print("\n✅ Todos los archivos copiados al RAW.")
```
