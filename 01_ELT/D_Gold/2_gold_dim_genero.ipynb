```python
spark.sql("DROP TABLE IF EXISTS workspace.gold.dim_genre")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS workspace.gold.dim_genre (
  genre_id BIGINT,
  genre_name STRING
)
""")
```

```python
df = spark.table("workspace.silver.movie_genres").toPandas()

dim_genre = (
    df.reindex(columns=["genre_id", "genre_name"])
    .drop_duplicates(subset=["genre_id"])
)

df_spark = spark.createDataFrame(dim_genre)
```

```python

# Usamos overwrite para reemplazar completamente los datos existentes
df_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.dim_genre")
```
