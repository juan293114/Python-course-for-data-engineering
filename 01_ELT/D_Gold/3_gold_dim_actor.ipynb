```python
spark.sql("DROP TABLE IF EXISTS workspace.gold.dim_actor")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS workspace.gold.dim_actor (
  actor_id BIGINT,
  actor_name STRING
)
""")
```

```python
df = spark.table("workspace.silver.movie_actors").toPandas()

dim_actor = (
    df.reindex(columns=["actor_id", "actor_name"])
    .drop_duplicates(subset=["actor_id"])
)

df_spark = spark.createDataFrame(dim_actor)
```

```python
df_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.dim_actor")
```
