DIM MOVIE

```python
spark.sql("DROP TABLE IF EXISTS workspace.gold.dim_movie")

spark.sql("""
CREATE TABLE IF NOT EXISTS workspace.gold.dim_movie (
  movie_id BIGINT,
  title STRING,
  original_title STRING,
  language STRING
)
""")
```

```python

df = spark.table("workspace.silver.movies").toPandas()
```

```python
dim_movie = (
    df.reindex(columns=[
        "movie_id",
        "title",
        "original_title",
        "language"
    ])
    .drop_duplicates(subset=["movie_id"])
)

df_spark = spark.createDataFrame(dim_movie)
```

```python

df_spark = spark.createDataFrame(dim_movie)

# Usamos overwrite para reemplazar completamente los datos existentes

df_spark.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.gold.dim_movie")
```
