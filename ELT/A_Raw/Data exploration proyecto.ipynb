```python
import requests
import pandas as pd
url = "https://www.datos.gov.co/resource/7y2j-43cv.json"

for year in range (2018, 2024):
    params = {
        "$where": f"year_radica='{year}'",
        "$limit": 15000
    }

    r = requests.get(url, params=params, timeout=30)
    print("Año:", year)
    print("Estado:", r.status_code)

    #print(r.json())
```

```python
df = pd.json_normalize(r.json())

df.head()
```
