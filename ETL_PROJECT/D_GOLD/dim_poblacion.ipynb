```python
from pyspark.sql.functions import *
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS workspace.gold;
```

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.gold.dim_poblacion (
    id_ubigeo STRING COMMENT 'Código único de distrito según INEI',
    poblacion INT COMMENT 'Población total del distrito al año 2025',
    fecha_actualizacion DATE COMMENT 'Fecha de actualización de la tabla Gold'
) USING DELTA;
```

```python
dim_poblacion = (
    spark.table("workspace.silver.m_ubigeo_demografia")
    .select(
            "id_ubigeo", 
            "poblacion"
    )
    .withColumn("fecha_actualizacion", current_date())
)

dim_poblacion.write.format("delta").mode("overwrite").saveAsTable("workspace.gold.dim_poblacion")
```
