```python
dbutils.widgets.text("api_token", "")
api_token = dbutils.widgets.get("api_token")
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from pyspark.sql import SparkSession

# Inicializar Spark
spark = SparkSession.builder.getOrCreate()

# Coordenadas de Lima
lat = -12.0464
lon = -77.0428
api_key = api_token

# Rango de fechas: últimos 30 días
end_date = int(datetime.now().timestamp())
start_date = int((datetime.now() - timedelta(days=30)).timestamp())

url = f"http://api.openweathermap.org/data/2.5/air_pollution/history?lat={lat}&lon={lon}&start={start_date}&end={end_date}&appid={api_key}"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    # Normalizar la lista de mediciones
    pdf = pd.json_normalize(data["list"])
    
    # Mostrar DataFrame pandas
    print(pdf.head())
    
    # Convertir a Spark DataFrame si lo necesitas
    df = spark.createDataFrame(pdf)
    df.display()
else:
    print(f"Error {response.status_code}: {response.text}")
```
