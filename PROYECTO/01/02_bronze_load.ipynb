```python
# Crear esquema bronze
spark.sql("CREATE DATABASE IF NOT EXISTS bronze")

# Leer desde raw
df_clientes_raw = spark.table("raw.clientes")
df_productos_raw = spark.table("raw.productos")
df_ordenes_raw = spark.table("raw.ordenes")
df_detalle_raw = spark.table("raw.detalle_orden")

# Guardar en bronze
df_clientes_raw.write.mode("overwrite").saveAsTable("bronze.clientes")
df_productos_raw.write.mode("overwrite").saveAsTable("bronze.productos")
df_ordenes_raw.write.mode("overwrite").saveAsTable("bronze.ordenes")
df_detalle_raw.write.mode("overwrite").saveAsTable("bronze.detalle_orden")

# Validar
spark.sql("SHOW TABLES IN bronze").show()
```
