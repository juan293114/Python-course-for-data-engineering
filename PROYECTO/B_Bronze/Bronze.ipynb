```python

from pyspark.sql.functions import input_file_name, current_timestamp

raw_calendar_path = "/Volumes/workspace/default/filestore/tables/raw/calendar.csv"

df_calendar_raw = (
    spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(raw_calendar_path)
        .withColumn("ingestion_timestamp", current_timestamp())


)

raw_products_path = "/Volumes/workspace/default/filestore/tables/raw/products.csv"

df_products_raw = (
    spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(raw_products_path)
        .withColumn("ingestion_timestamp", current_timestamp())


)


raw_customers_path = "/Volumes/workspace/default/filestore/tables/raw/customers.csv"

df_customers_raw = (
    spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(raw_customers_path)
        .withColumn("ingestion_timestamp", current_timestamp())


)


raw_stores_path = "/Volumes/workspace/default/filestore/tables/raw/stores.csv"

df_stores_raw = (
    spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(raw_stores_path)
        .withColumn("ingestion_timestamp", current_timestamp())


)

raw_sales_path = "/Volumes/workspace/default/filestore/tables/raw/sales.csv"

df_sales_raw = (
    spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(raw_sales_path)
        .withColumn("ingestion_timestamp", current_timestamp())


)



delta_raw_calendar_path = "/Volumes/workspace/default/filestore/tables/delta_raw/raw_calendar/"
df_calendar_raw.write.format("delta").mode("overwrite").save(delta_raw_calendar_path)

delta_raw_customers_path = "/Volumes/workspace/default/filestore/tables/delta_raw/raw_customers/"
df_calendar_raw.write.format("delta").mode("overwrite").save(delta_raw_customers_path)

delta_raw_products_path = "/Volumes/workspace/default/filestore/tables/delta_raw/raw_products/"
df_calendar_raw.write.format("delta").mode("overwrite").save(delta_raw_products_path)

delta_raw_stores_path = "/Volumes/workspace/default/filestore/tables/delta_raw/raw_stores/"
df_stores_raw.write.format("delta").mode("overwrite").save(delta_raw_stores_path)

delta_raw_sales_path = "/Volumes/workspace/default/filestore/tables/delta_raw/raw_sales/"
df_calendar_raw.write.format("delta").mode("overwrite").save(delta_raw_sales_path)
```
