```python
%sql
CREATE DATABASE IF NOT EXISTS workspace.silver
COMMENT 'Capa Silver procesados'
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS workspace.silver.uses (
  id BIGINT,
  fechaCreacion TIMESTAMP,
  terminal_id BIGINT,
  tsn BIGINT,
  veh_id BIGINT,
  line_id BIGINT,
  app_id BIGINT,
  type_use BIGINT,
  fee BIGINT,
  uid BIGINT,
  trx_date TIMESTAMP,
  flee_control STRING,
  state BIGINT,
  state_description STRING,
  tp_Id BIGINT,
  purseavalue DECIMAL(7,2),
  pursebvalue DECIMAL(7,2),
  cut_id INT,
  errorcode BIGINT
)
USING DELTA
""")
```

```python
query = f"select max(a.id) from workspace.bronze.uses a;"
max_id = 112451000 #spark.sql(query).first()[0]
print(max_id)
```

```python
df_spark = spark.sql(f"""select *
from workspace.bronze.uses a
where a.id > {max_id}""")
df = df_spark.toPandas()
```

```python
#display(df_spark.limit(5))
```

```python
rename_map = {"validator_id":"terminal_id",
              "bus_id":"veh_id",
              "applic_id":"app_id",
              "uid_target":"uid",
              "date":"trx_date",
}

df = df.rename(columns=rename_map, errors="ignore")
df.head(5)
```

```python
# 6) Quitar duplicados
df = df.drop_duplicates()

# 9) Regresar a Spark DataFrame
df_spark = spark.createDataFrame(df)

#display(df_spark.limit(5))
```

```python
# Usamos append para reemplazar completamente los datos existentes
df_spark.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("append") \
    .saveAsTable("workspace.silver.uses")

```
